```python
import re
import numpy  as np
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import plotly.colors as colors
from wordcloud import WordCloud
from sklearn.neighbors import NearestNeighbors
import warnings
warnings.filterwarnings("ignore")
```

```python
books_data = pd.read_csv("data/books.csv", on_bad_lines='skip')
books_data.head()
```

```python
print(books_data.columns)
# rename columns
books_data = books_data.rename(columns={
    'bookID': 'book_id',
    'title': 'title',
    'authors': 'authors',
    'average_rating': 'average_rating',
    'isbn': 'isbn',
    'isbn13': 'isbn13',
    'language_code': 'language_code',
    '  num_pages': 'num_of_pages',
    'ratings_count': 'ratings_count',
    'text_reviews_count': 'text_reviews_count',
    'publication_date': 'publication_date',
    'publisher': 'publisher'
})
```

```python
books_data.info()
```

```python
print(f"dataset contains {books_data.shape[0]} rows and {books_data.shape[1]} columns")
```

```python
# handling missing values
print(books_data.isnull().sum())  # Identify missing values

books_data['publication_date'] = books_data['publication_date'].fillna('Unknown')
```

```python
# Convert Data Types: Ensure numeric and date fields are properly formatted:
books_data['average_rating'] = pd.to_numeric(books_data['average_rating'], errors='coerce')
books_data['ratings_count'] = pd.to_numeric(books_data['ratings_count'], errors='coerce')
books_data['publication_date'] = pd.to_datetime(books_data['publication_date'], errors='coerce')
```

```python
# Author based aggregations
author_stats = books_data.groupby('authors').agg({
    'book_id': 'count',                 # Number of books published
    'average_rating': 'mean',          # Average rating of books
    'ratings_count': 'sum',            # Total number of ratings
    'text_reviews_count': 'sum'        # Total number of reviews
}).rename(columns={
    'book_id': 'total_books',
    'average_rating': 'avg_rating',
    'ratings_count': 'total_ratings',
    'text_reviews_count': 'total_reviews'
}).reset_index()

print(author_stats.head())
```

```python
#identify High-Performing Authors: average rating above 4.0 or more than 10 books (subject to change)
high_performers = author_stats[
    (author_stats['avg_rating'] >= 4.0) &
    (author_stats['total_books'] > 10)
]
print(high_performers)
```

```python
# Visualizing what we have for now
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(
    data=author_stats, 
    x='total_books', 
    y='avg_rating', 
    size='total_ratings', 
    hue='avg_rating', 
    palette='coolwarm'
)
plt.title("Number of Books vs Average Rating")
plt.xlabel("Total Books Published")
plt.ylabel("Average Rating")
plt.show()

# TODO: resolving the abnormal values
```

```python
top_authors = books_data['authors'].value_counts().head(10).index
sns.boxplot(
    data=books_data[books_data['authors'].isin(top_authors)], 
    x='authors', 
    y='average_rating'
)
plt.title("Rating Distribution for Top 10 Authors")
plt.xticks(rotation=45)
plt.show()
```
